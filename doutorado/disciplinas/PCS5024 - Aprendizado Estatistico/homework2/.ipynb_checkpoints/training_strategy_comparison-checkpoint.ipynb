### This notebook contains a simple example of how two different RNN training regimes could be compared.

1. Decoder: Only decoder outputs are used for training
2. Encoder+Decoder: Both encoder and decoder outputs are used for training. Test phase still only considers decoder losses.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
from torch.utils.data import TensorDataset, DataLoader
import gdown
from tqdm import tqdm
import uniplot
```

```python
id = "1qZv6wwHLyMIZQQNQN8NN676AIgF-XJt5"
gdown.download(id=id, output="santos_ssh.csv", quiet=False)

df = pd.read_csv("santos_ssh.csv")
df["datetime"] = pd.to_datetime(df["datetime"])
train_df = df[df["datetime"].dt.tz_convert(None) < np.datetime64("2020-06-01 00:00:00")]
test_df = df[df["datetime"].dt.tz_convert(None) >= np.datetime64("2020-06-01 00:00:00")]
train_df.head(), test_df.head()
```

```python
train_df["ssh"].iloc[:1000].plot()
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
input_size = 1
hidden_size = 64
num_epochs = 30
past_len = 800
future_len = 500
batch_size = 32
learning_rate = 1e-3
device
```

```python
# Here we write a SimpleARModel that uses the network 
# output as the input for the next prediction step (i.e. no teacher forcing)
class SimpleARModel(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.rnn = nn.GRU(input_size, hidden_size, batch_first=True)
        self.linear = nn.Linear(hidden_size, input_size)

    def forward(self, x, target_seq_len):
        # encoding
        out, h_n = self.rnn(x)
        encoder_out_proj = self.linear(out.reshape(-1, self.hidden_size)).view(
            x.shape[0], -1, self.input_size
        ) # project the output of the encoder to the input size
        inp = encoder_out_proj[:, -1]

        output_seq = torch.empty(
            x.shape[0], target_seq_len, x.shape[-1], device=inp.device
        ) # allocate memory for the output sequence
        output_seq[:, 0] = inp

        for i in range(1,target_seq_len):
            out, h_n = self.rnn(inp.unsqueeze(1), h_n)
            inp = self.linear(out[:, -1])
            output_seq[:, i] = inp
        return encoder_out_proj[:,:-1],output_seq # last decoder proj is the first element of the output_seq


def create_sequences(data, past_len, future_len):
    xs, ys = [], []
    for i in range(past_len, len(data) - future_len):
        x = data[(i - past_len) : i]
        y = data[i : (i + future_len)]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)
```

```python
train_data = train_df["ssh"].values
test_data = test_df["ssh"].values

X_train, y_train = create_sequences(train_data, past_len, future_len)
X_test, y_test = create_sequences(test_data, past_len, future_len)

X_train = torch.from_numpy(X_train.astype(np.float32))
y_train = torch.from_numpy(y_train.astype(np.float32))
X_test = torch.from_numpy(X_test.astype(np.float32))
y_test = torch.from_numpy(y_test.astype(np.float32))

train_dataset = TensorDataset(X_train, y_train)
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

test_dataset = TensorDataset(X_test, y_test)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
def run_epoch(model, criterion, optimizer,epoch,use_encoder_loss):
    model.train()
    for inputs, targets in tqdm(train_dataloader):
        inputs = inputs.unsqueeze(-1).to(device)
        targets = targets.unsqueeze(-1).to(device)

        encoder_proj,outputs = model(inputs, targets.shape[1])
        if use_encoder_loss:
            loss = criterion( # train using both encoder and decoder outputs
                torch.cat([encoder_proj,outputs],dim=1),
                torch.cat([inputs[:,1:],targets],dim=1) # first input is not a label to any encoder output
                )
        else:
            loss = criterion(outputs, targets)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    model.eval()
    with torch.no_grad():
        test_losses = []
        for i, (inputs, targets) in enumerate(tqdm(test_dataloader)):
            inputs = inputs.unsqueeze(-1).to(device)
            targets = targets.unsqueeze(-1).to(device)

            _,outputs = model(inputs, targets.shape[1])
            test_loss = criterion(outputs, targets) # but evaluate the model only on decoder outputs
            test_losses.append(test_loss.item())
            past_view_size = 50
            if i ==0:
                inp = inputs[0, :, 0].cpu()
                tar = targets[0, :, 0].cpu()
                out = outputs[0, :, 0].cpu()
                uniplot.plot(
                    ys=[inp[-past_view_size:], tar, out],
                    xs=[
                        np.arange(0, past_view_size),
                        np.arange(past_view_size, past_view_size + len(tar)),
                        np.arange(past_view_size, past_view_size + len(tar)),
                    ],
                    color=True,
                    legend_labels=["Input", "Target", "Output"],
                    title=f"Epoch: {epoch}, Test Loss: {test_loss.item():.4f} - Example {i} - Use Encoder Loss: {use_encoder_loss}",
                    height=10
                )

    mean_test_loss = np.mean(test_losses)
    return mean_test_loss
```

```python
def run_experiment(models,criterion,optimizers):
    test_losses = {t:[] for t in models.keys()}
    for epoch in range(1,num_epochs+1):
        for use_encoder_loss,model in models.items():
            test_loss = run_epoch(model,criterion,optimizers[use_encoder_loss],epoch,use_encoder_loss)
            test_losses[use_encoder_loss].append(test_loss)
        uniplot.plot(
                ys=[test_losses[True],test_losses[False]],
                xs=[np.arange(1,epoch+1),np.arange(1,epoch+1)],
                color=True,
                legend_labels=["Encoder+Decoder", "Decoder only"],
                title=f"Epoch: {epoch}, Test Losses.",
                height=10
            )
    return test_losses
```

```python
models = {t:SimpleARModel(input_size, hidden_size).to(device) for t in [True,False]}
criterion = nn.MSELoss()
optimizers = {t:optim.Adam(models[t].parameters(), lr=learning_rate) for t in models.keys()}


test_losses = run_experiment(models,criterion,optimizers)

```

```python
import matplotlib.pyplot as plt
N = 5
run_mean = lambda x, N: np.convolve(x, np.ones((N,))/N, mode='valid')
plt.plot(run_mean(test_losses[True],N),label="Encoder+Decoder")
plt.plot(run_mean(test_losses[False],N),label="Decoder only")
plt.legend()
plt.title("Test Losses")
plt.show()
```

### Conclusion:
1. Decoder only strategy converged slightly more quickly
2. Encoder+Decoder strategy was more resistant to overfitting
3. Both strategies reached similar minimum values
